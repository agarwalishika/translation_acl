```python
from datasets import load_dataset

ds = load_dataset("Helsinki-NLP/opus-100", "en-zh")['test'].to_pandas()
```

```python
ds
```

```python
# src, true_meaning, literal_translation
df = {"src": [], "true_meaning": []}

for i, row in ds.iterrows():
    if i >= 400: break
    df['src'].append(row['translation']['zh'])
    df['true_meaning'].append(row['translation']['en'])

import pandas as pd
df = pd.DataFrame.from_dict(df)
```

```python
df.to_csv('opus_chinese.csv')
```
